```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("population-and-demography.csv")

country_name = [0, 0, 0]


country_name[0] = input("Enter the country you want to find")
country_name[1] = input("Enter the country you want to find")


country_data = []

country_data.append(df[df["Country name"] == country_name[0]])
country_data.append(df[df["Country name"] == country_name[1]])

plt.figure(figsize=(10,10))
sns.scatterplot(data = country_data[0], x="Year", y="Population", label=country_name[0])
sns.scatterplot(data = country_data[1], x="Year", y="Population", label=country_name[1])


# Cubic    
X = country_data[0]["Year"]
y = country_data[0]["Population"]
a0, a1, a2, a3= np.polyfit(x=X, y=y, deg=3)


X = country_data[1]["Year"]
y = country_data[1]["Population"]
b0, b1, b2, b3= np.polyfit(x=X, y=y, deg=3)

num_of_year = int(input("Enter the number of years to predict: "))
prediction_year = np.linspace(1950, (num_of_year+2021), (num_of_year+73))

population_prediction = a0*prediction_year*prediction_year*prediction_year + a1*prediction_year*prediction_year + a2*prediction_year + a3
plt.plot(prediction_year, population_prediction, color = "blue")

population_prediction = b0*prediction_year*prediction_year*prediction_year + b1*prediction_year*prediction_year + b2*prediction_year + b3
plt.plot(prediction_year, population_prediction, color = "orange")

# Quadratic
X = country_data[0]["Year"]
y = country_data[0]["Population"]
a0, a1, a2= np.polyfit(x=X, y=y, deg=2)


X = country_data[1]["Year"]
y = country_data[1]["Population"]
b0, b1, b2= np.polyfit(x=X, y=y, deg=2)

prediction_year = np.linspace(1950, (num_of_year+2021), (num_of_year+73))

population_prediction = a0*prediction_year*prediction_year + a1*prediction_year + a2
plt.plot(prediction_year, population_prediction, color = "blue", linestyle='dashed')

population_prediction = b0*prediction_year*prediction_year + b1*prediction_year + b2
plt.plot(prediction_year, population_prediction, color = "orange", linestyle='dashed')
```
